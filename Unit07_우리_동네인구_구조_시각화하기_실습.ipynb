**우리 동네 연령별 인구수 result 리스트에 저장하기 -> 시각화를 위해 **

```python
import csv

f = open('/content/age.csv', encoding= 'cp949')

data =csv.reader(f)

result = [] # 0세 부터 100 시각화 하기 위해서 빈 리스트에 담음
#row[0]: 지역명, row[1]: 총인구수, row[2]: 연령구간인구수, row[3]~  인덱스 부터 연령별 인구 수

for row in data:
  if '아름동' in row[0]:
    for i in row[3:] :
      result.append(int(i.replace(',',''))) #인구수 데이터 정수로 변환 후 result 변수에 담기 #천명이 넘으면 리플레이스 사용 아니면INT사용가능

print(result)
```

**우리 동네 연령별 인구수 데이터 시각화하**

```python
import matplotlib.pyplot as plt

plt.figure(dpi = 130)
plt.style.use('Solarize_Light2') #ggplot 격자 무늬 스타일로 지정# 격자 무늬 스타일로 지정 classic, dark_background

plt.plot(result)
plt.show() 
# 아름동 학원가 반경 1km 안에 초중고가 다있고  연령층이 낮다. 십대가 많기 때문에 부모의 연령도 낮다 때문에 십대 및 30~40대가 많음
```
