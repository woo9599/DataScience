**[step1] 데이터 준비 - seaborn에서 제공하는 titanic 데이터셋 가져오**

```python
# load_dataset 함수를 사용하여 데이터프레임으로 변환

df = sns.load_dataset('titanic')
```

**[step2] 데이터 탐색 및 데이터 전처리**



```python
df.head()
```

```python
df.info()
```

```python
#naa값이 만은 deck(객실데크위치) 열을 삭제, embarked와 내용이 겹치는 embark_town(승선도시)열삭제

rdf = df.drop(['deck', 'embark_town'], axis = 1)
rdf.info()
```

```python
#age 열에 나이 데이터가 없는 모든행을 삭제 = age열(891개 중 177개의  nan값)
rdf = rdf.dropna(subset = ['age'], how = 'any', axis = 0)

rdf.info()
```

```python
# embarked 열의 NaN값을 승선도시 중에서 가장 많이 출현한 값 확인 -> 탑승한 도시의 첫글자

most_freq = rdf['embarked'].value_counts(dropna = True).idxmax()
most_freq
```

```python
# embarked 열의 NaN값을 승선도시 중에서 가장 많이 출현한 값으로 치환하기


rdf['embarked'].fillna(most_freq, inplace = True)
```

**[setp]분석에 사용할 속성을 선택**

```python
#분석에 활용할 열(속성)을 선택 (생존여부, 객실 등급,성별,나이,타이타닉에 탑승한 형제 /자매수,)
#타이타닉에 탑승한 부모/자녀수 , 탑승한곳(항구) c = cherbourg, q = queenstown, s = southampton)

ndf = rdf[['survived', 'pclass','sex','age','sibsp','parch','embarked']]
ndf.head()
```

```python
#embarked c ,q ,s->[1,0,0],[0,1,0] , [0,0,1]

onehot_sex = pd.get_dummies(ndf['sex'])
ndf =pd.concat([ndf, onehot_sex], axis=1)

ndf.head()
```

```python
#원핫인코딩 - 범주형 데이터를 모형이 인식할수 있도록 숫자형으로 변경

# male, female -> [1,0],[0,1]
#embarked c ,q ,s->[1,0,0],[0,1,0] , [0,0,1]
onehot_embarked = pd.get_dummies(ndf['embarked'],prefix='town')
ndf =pd.concat([ndf, onehot_embarked], axis=1)

ndf.head()
```

[step4] 데이터셋 구분- 훈령용(train data)/검증용(test data)

```python
# ndf -> x,y (문제집, 정답지)

X = ndf[['pclass', 'age', 'sibsp', 'parch', 'female', 'male','town_C', 'town_Q', 'town_S']] # 독립변수
y = ndf['survived']
```

```python
#설명 변수(데이터,문제집) 정규화(normalization)-> 스케일링 0~1사이의 작은 값으로 축소
from sklearn import  preprocessing

X= preprocessing.StandardScaler().fit(X).transform(X)

X
```

```python
# train 데이터와 test 데이터로 구분(7:3 비율)

from sklearn.model_selection import train_test_split

X_train,X_test, y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state =10)

print('train data 개수:',X_train.shape)
print('test data 개수:',X_test.shape)
```

[step 5] SVM분류 모형- sklearn 사용 

```python
# sklearn 라이브러리에서 svm 분류 모형 가져오기

from sklearn import svm

# 모형 객체 생성 (kernel='rbf' 적용)
# 벡터 공간을 맵핑 하는 함수 -> radial basis function (RBF) 커널을 사용
# 선형(linear), 다항식(poly), 가우시안 RBF(rbf) , 시그모이드(sigmoid)

svm_model = svm.SVC(kernel ='rbf')# 벡터 공간을 맵핑 하는 함수 -> 기본 rbf 사용
```

```python
# 훈련 데이터를 사용하여 모형학습
svm_model.fit(X_train, y_train)

#테스트 데이터를 사용하여 y_hat 예측ㄹ
y_hat = svm_model.predict(X_test)
```

**[step6] 모형 성능 평가**

```python
from sklearn import metrics

print("테스트 세트 정확도 :{:.2f}%".format(svm_model.score(X_test,y_test)*100))
```

```python
#모형 성능평가 평가지표 계산

svm_report = metrics.classification_report(y_test,y_hat)
print(svm_report)
```
