**최고 기온 데이터 히스토그램으로 표현하기**

**1월과 8월의 데이터를 히스토그램으로 시각화 하기**

```python
import csv
import matplotlib.pyplot as plt


f = open('/content/seoul.csv',encoding= 'cp949')

data = csv.reader(f)
next(data) #헤더

Aug = [] # 8월의 최고 기온을 담을 aug 리스트 변수 선언
Jan = [] # 1월의 최고 기온을 담을 jan 리스트 변수 선언
june = [] # 6월의 최고 기온을 담을 june 리스트 변수 선언

for row in data:
  month = row[0].split('-')[1] #날짜 데이터에서 월 데이터 추출
  
  if row[-1] != '':
    if month == '08': # 8월 데이터면 aug 변수에 담기
      Aug.append(float(row[-1]))
    if month == '01': # 1월 데이터면 jan 변수에 담기
     Jan.append(float(row[-1]))
    if month == '04': # 1월 데이터면 june 변수에 담기
     june.append(float(row[-1]))


# 최고기온 데이터 히스토그램
plt.figure(dpi=200)
plt.hist(Aug,bins=100, color ='b', label="Aug") #result
plt.hist(Jan, bins =100, color='r', label="Jan")
plt.hist(june, bins =100, color='k', label="June")
plt.legend() 
plt.show()
```

**1월 부터 12월 까지 최고 기온 데이터 상자 그림으로 표현하기**

```python
import csv
import matplotlib.pyplot as plt


f = open('/content/seoul.csv',encoding= 'cp949')

data = csv.reader(f)
next(data) #헤더

day = [] # 일별 데이터를 저장할 리스트 day 변수 선언
for i in range(12) :
  day.append([]) #day 리스트 내에 12개 리스트 생성([],[],[],[]...31개)

#month = [[],[],[],[],[],[],[],[],[],[],[],[]]# 월별 데이터 저장할 month변수 선언

for row in data:
  if row[-1] != '':
    month[int(row[0].split('-')[1])-1].append(float(row[-1])) # 월과 같은 번호의 인덱스에 월별 데이터 저장 (예: 1월 row[0].split('-')[1] -> month[0] row[0].split('-')[1]-1 , 2월 -> month[1])

    
plt.figure(dpi=200)
plt.boxplot(month)
plt.show()
```

**8월1일부터 31일까지 최고 기온 데이터 상자 그림으로 표현하기**

```python
import csv
import matplotlib.pyplot as plt


f = open('/content/seoul.csv',encoding= 'cp949')

data = csv.reader(f)
next(data) #헤더

day = [] # 일별 데이터를 저장할 리스트 day 변수 선언
for i in range(31) :
  day.append([]) #day 리스트 내에 31개 리스트 생성([],[],[],[]...31개)


for row in data:
  if row[-1] != '':
    if row[0].split('-')[1] == '08' : # 8월 최고 기온 값 저장 day[0] -> 1일
      day[int(row[0].split('-')[2])-1].append(float(row[-1]))

plt.figure(figsize =(10,5), dpi=200)
plt.boxplot(day, showfliers=False) # 아웃라이어(이상치) 값 생략
plt.show()
```
