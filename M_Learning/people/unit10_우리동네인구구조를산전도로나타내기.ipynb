**꺾은선 그래프로 표현하기**

```python
import csv
import matplotlib.pyplot as plt



f = open('/content/gender.csv', 'r', encoding = 'cp949')
data = csv.reader(f)

m = []
f = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
  if name in row[0]:
    for i in row[3:104]:    # 남성 : 3 ~ 103번 데이터 슬라이싱 - 정수 데이터로 변환 - 리스트에 저장
      m.append(int(i.replace(',', '')))
    for b in row[106:]:
      f.append(int(b.replace(',', '')))      # 여성 : 106 ~ 끝까지 데이터 슬라이싱 ''
    break 

plt.figure(dpi = 150)
plt.rcParams['axes.unicode_minus'] = False # 숫자,문자 오류처리

#꺾은선 그래프로 남여 인구수 데이터 표현하기
plt.plot(m, label = '남성')
plt.plot(f, color = 'red',label = '여성')
plt.legend()
plt.show()

```

**연령별로 남성 인구수에서 여성 인구수를 뺀 차이의 막대그래프 그리기
**

```python
import csv
import matplotlib.pyplot as plt



f = open('/content/gender.csv', 'r', encoding = 'cp949')
data = csv.reader(f)

result = [] # 남성인구 - 여성 인구 값 담을 리스트 변수 선언

name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
  if name in row[0]:
    for i in range(3,104):
      result.append(int(row[i].replace(',', '')) - int(row[i+103].replace(',', '')))
    break


plt.rc('font', family = 'NanumBarunGothic') # 나눔 고딕
plt.rcParams['axes.unicode_minus'] = False # 숫자,문자 오류처리
plt.figure(dpi = 150)
plt.title(name+'지역의 남녀 성별 인구 차이')
plt.bar(range(101),result)


plt.show()
```

**위치,크기가 서로 다른 100개의 점 만들기**

**원의 크기는 인구수의 제곱근 추가**

```python
import csv
import matplotlib.pyplot as plt
import math


f = open('/content/gender.csv', 'r', encoding = 'cp949')
data = csv.reader(f)

m = []
f = []
size =[] # 인구수 사이즈 리스트 변수 선언 

name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
  if name in row[0]:
    for i in row[3:104]:    # 남성 : 3 ~ 103번 데이터 슬라이싱 - 정수 데이터로 변환 - 리스트에 저장
      m.append(int(i.replace(',', '')))
      size.append(math.sqrt(int(i.replace(',', ''))))
    for b in row[106:]:
      f.append(int(b.replace(',', '')))
    break 



#인구수와 sqrt(인구수) 차이 확인
print(m)
print(size)

#산점도 그래프로 남여 인구수 데이터 표현하기
plt.figure(dpi = 150)
plt.scatter(m,f,s = size,c =range(101),cmap ='jet',alpha=0.5) # x축은 남자 y축은 여자 
plt.plot(range(max(m)),range(max(m)),'g')
plt.colorbar()
plt.show()
```
