**제주특별자치도의 성별 분포를 항아리 모양 그래프로 표현 **

```python
import csv
import matplotlib.pyplot as plt



f = open('/content/gender.csv', 'r', encoding = 'cp949')
data = csv.reader(f)

m = []
f = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
  if name in row[0]:
    for i in row[3:104]:    # 남성 : 3 ~ 103번 데이터 슬라이싱 - 정수 데이터로 변환 - 리스트에 저장
      m.append(-int(i.replace(',', '')))
    for b in row[106:]:
      f.append(int(b.replace(',', '')))      # 여성 : 106 ~ 끝까지 데이터 슬라이싱 ''
    break   # 첫번 째 찾고 for문 중단


plt.rc('font', family = 'NanumBarunGothic') # 나눔 고딕
plt.rcParams['axes.unicode_minus'] = False # 숫자,문자 오류처리
plt.figure(dpi = 150)
plt.title(name+'지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label = '남성')
plt.barh(range(101), f, color ='red', label = '여성')
plt.legend()
plt.show
```

**제주특별자치도 데이터확인 -> 101개여야 하는데,4646 -> 46배 큰값 -> 제주특별자치도 포함된 지역 이름이 46건**

**제주도의 성별 인구 비율 표현하기 **

```python
import csv
import matplotlib.pyplot as plt



f = open('/content/gender.csv', 'r', encoding = 'cp949')
data = csv.reader(f)

size = [] # 남제주도의 남성 인구 합계, 여성 인구 합계


name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
  if name in row[0]:
    m = 0 # 남자 합계 인구 변수
    f = 0 # 여자 합계 인구 변수
    for i in range(101) :
      m += int(row[3+i].replace(',','')) # m= m +int(row[3+i].replace(',','')) 
      f += int(row[106+i].replace(',',''))   # f= f +int(row[3+i].replace(',',''))  
    break   

print(m)
print(f)


size.append(m)
size.append(f)



#파이 차트 시각화
plt.figure(dpi = 150)
color =[]
color = ['crimson', 'darkcyan']
label =['남','여']
plt.axis('equal')
plt.pie(size, labels=label,autopct='%.1f%%',explode =(0,0.1),colors = color,startangle=270)
plt.title(name + '지역의 남녀 성별 비율')
plt.legend() # loc = 
plt.show()


```
