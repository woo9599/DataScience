**지역명을 입력받아서 연령별 인구수 데이터 시각화하기**

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/age.csv', 'r', encoding = 'cp949')
data = csv.reader(f)

result = []   # 순서를 위해 인덱스가 있는 리스트로 담는다.

name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')

for row in data:
  if name in row[0]:
    for i in row[3:]:
      result.append(int(i.replace(',', '')))   # 인구수 데이터 정수로 변환 후 result 변수에 저장

# 인구구조 시각화
plt.figure(dpi = 110)
plt.title(name+' 지역의 인구 구조')
plt.style.use('ggplot')
plt.plot(result)
plt.show()
```

**막대 그래프 그리기 -> 비교(지역별,성별 기준등)**

```python
import csv

f = open('/content/age.csv', encoding= 'cp949')

data =csv.reader(f)

result = [] # 0세 부터 100 시각화 하기 위해서 빈 리스트에 담음
#row[0]: 지역명, row[1]: 총인구수, row[2]: 연령구간인구수, row[3]~  인덱스 부터 연령별 인구 수

for row in data:
  if '아름동' in row[0]:
    for i in row[3:] :
      result.append(int(i.replace(',',''))) #인구수 데이터 정수로 변환 후 result 변수에 담기 #천명이 넘으면 리플레이스 사용 아니면INT사용가능

plt.figure(dpi = 130)
plt.bar(range(101),result)
plt.show()
```

**항아리 모양 그래프 그리기**

성별데이터 저장하기(type1)

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/gender.csv', encoding= 'cp949')

data =csv.reader(f)

m = []
f = []

for row in data:
  if '신도림' in row[0]:
    for i in range(0,101):
      m.append(int(row[i+3]))
      f.append(int(row[-(i+1)]))# -1 값부터 담음 -> 100세 이상 부터 값이 담김
     



f.reverse() #0세부터 인구수 나올수 있도록 데이터 뒤집기

print(m)
print(f)
```

**성별 데이터 순서대로 저장하기(Type2)**

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/gender.csv', encoding= 'cp949')

data =csv.reader(f)

m = []
f = []

for row in data:
  if '신도림' in row[0]:
    for i in row[3:104]: # 남성 데이터 : 3~ 103번 데이터 슬라이싱 - 정수 데이터로 변환 - m리스트에 저장
      m.append(int(i))
    for i in row[106:]: # 여성 데이터 : 106~ 끝까지 데이터 슬라이싱 - 정수 데이터로 변환 - f리스트에 저장
      f.append(int(i))

print(m)
print(f)
```

**데이터 시각화 하기**

```python
import matplotlib.pyplot as plt

plt.figure(dpi = 110)
plt.style.use('ggplot')
plt.barh(range(101),m)
plt.barh(range(101),f)
plt.show()
```

**항아리 그래프를 위해 남성 데이터 음수로 바꾸기**

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/gender.csv', encoding= 'cp949')

data =csv.reader(f)

m = []
f = []

for row in data:
  if '아름동' in row[0]:
    for i in row[3:104]: # 남성 데이터 : 3~ 103번 데이터 슬라이싱 - 정수 데이터로 변환 - m리스트에 저장
      m.append(-int(i)) # 남성 데이터 음수로 변경
    for i in row[106:]: # 여성 데이터 : 106~ 끝까지 데이터 슬라이싱 - 정수 데이터로 변환 - f리스트에 저장
      f.append(int(i))

print(m)
print(f)
```

```python
import matplotlib.pyplot as plt

plt.figure(dpi = 110)
plt.rc('font', family = 'NanumBarunGothic') # 나눔 고딕
plt.rcParams['axes.unicode_minus'] = False # 숫자,문자 오류처리
plt.title('아름동 지역의 남녀 성별 인구 분포')
#plt.style.use('ggplot')
plt.barh(range(101),m,'label='dw')
plt.barh(range(101),f,label='dd')
plt.legend()
plt.show()
```
