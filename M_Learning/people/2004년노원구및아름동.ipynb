2) 2019년 2월 연령별 인구 구조 -> 2020년 2월 연령별 인구구조 비교 (같이 그리기) 

```python
import csv

f = open('/content/worldCup2004.csv', encoding= 'cp949')

data =csv.reader(f)

result = [] # 0세 부터 100 시각화 하기 위해서 빈 리스트에 담음
#row[0]: 지역명, row[1]: 총인구수, row[2]: 연령구간인구수, row[3]~  인덱스 부터 연령별 인구 수

for row in data:
  if '노원구' in row[1]:
    for i in row[4:] :
      result.append(float(i.replace(',',''))) #인구수 데이터 정수로 변환 후 result 변수에 담기 #천명이 넘으면 리플레이스 사용 아니면INT사용가능

print(result)
```

```python
import matplotlib.pyplot as plt

plt.figure(dpi = 130)
plt.style.use('Solarize_Light2') #ggplot 격자 무늬 스타일로 지정# 격자 무늬 스타일로 지정 classic, dark_background
plt.title('2004년 노원구 인구표')
plt.plot(result)
plt.show() 
```
