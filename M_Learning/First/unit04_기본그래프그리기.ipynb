**plt.plot(데이터셋) 그래프 그리기**

```python
import matplotlib.pyplot as plt # 라이브러리 임포트를 불러 온다

plt.plot([10,20,30,40]) # y값이 주어지면 , 자동으로 x 값을 0 부터 1씩 증가함 (우리가 찾고자하는 해)
plt.show() # 차트를 보여줘라 
```

**plt.plot(x,y) 그래프 그리기**

```python
import matplotlib.pyplot as plt # 1) 라이브러리 임포트를 불러 온다

plt.plot([1,2,3,4],[12,43,25,15]) # 2) plt.plot(x,y) 함수에 데이터 입력하기
plt.show() # 차트를 보여주기
```

```python
import matplotlib.pyplot as plt # 1) 라이브러리 임포트를 불러 온다

plt.title('plotting') # 타이틀 속성 추가
plt.plot([1,2,3,4],[12,43,25,15]) # 2) plt.plot(x,y) 함수에 데이터 입력하기
plt.show() # 차트를 보여주기
```

```python
import matplotlib.pyplot as plt # 1) 라이브러리 임포트를 불러 온다

plt.title('plotting') # 타이틀 속성 추가

plt.plot([10,20,30,40],label = 'asc') # 증가를 의미하는 asc 범례
plt.plot([40,30,20,10],label = 'desc') # 감소를 의미하는 asc 범례

plt.legend(loc =9) # 그래프에 범례 속성 추가
plt.show()
```

```python
import matplotlib.pyplot as plt # 1) 라이브러리 임포트를 불러 온다

plt.title('color') # 타이틀 속성 추가

plt.plot([10,20,30,40],color = 'skyblue',label = 'skyblue') # 하늘색 그래프 
plt.plot([40,30,20,10],'pink',label='pink') # 분홍색 그래프 

plt.legend(loc =9) # 그래프에 범례 속성 추가
plt.show()
```

```python
import matplotlib.pyplot as plt # 1) 라이브러리 임포트를 불러 온다

plt.title('color') # 타이틀 속성 추가

plt.plot([10,20,30,40],color = 'r',label = 'red') # 기본 색상은 약어로 표시 가능   
plt.plot([40,30,20,10],'g',label='green') # 초록색 그래프

plt.legend() # 그래프에 범례 속성 추가
plt.show()
```

```python
import matplotlib.pyplot as plt # 1) 라이브러리 임포트를 불러 온다

plt.title('linestyle') # 타이틀 속성 추가

plt.plot([10,20,30,40],color = 'r', linestyle = '--', label= 'dashed') # 빨간색dashed그래프 
plt.plot([40,30,20,10],'g',ls= ':',label='dotted') # 초록색 dotted 그래프 linestyle -> ls표현 가능 

plt.legend() # 그래프에 범례 속성 추가
plt.show()
```

```python
import matplotlib.pyplot as plt # 1) 라이브러리 임포트를 불러 온다

plt.title('color') # 타이틀 속성 추가

plt.plot([10,20,30,40],'r.',label = 'ciecle') # 빨간색 ciecle
plt.plot([40,30,20,10],'g^',label='triangle up') # 초록색 triangle up 그래프 linestyle -> ls표현가능

plt.legend() # 그래프에 범례 속성 추가
plt.show()
```

```python
import matplotlib.pyplot as plt # 1) 라이브러리 임포트를 불러 온다

plt.title('color') # 타이틀 속성 추가

#'r.--' : <색상><마커모양><선모양> 코드로 작성 가능 
plt.plot([10,20,30,40],'r.-',label = 'circle') # 빨간색 circle
plt.plot([40,30,20,10],'g^--',label='triangle up') # 초록색 triangle up 그래프 linestyle -> ls표현가능

plt.legend() # 그래프에 범례 속성 추가
plt.show()
```

**데이터시각화하기**

```python
import csv
import matplotlib.pyplot as plt# 라이브러리 임포트

f = open('/content/seoul.csv','r',encoding='cp949')
data = csv.reader(f)
header = next(data)

result =[] # 결측치를 제거한 최고 기온 리스트 

for row in data :
  if row[-1] != '' : # 최고 기온이 ''(빈문자열)이 아니면 
    result.append(float(row[-1]))
result

plt.figure(dpi = 200) #figsize=(10,2),
plt.plot(result, 'r')
plt.show()
```
